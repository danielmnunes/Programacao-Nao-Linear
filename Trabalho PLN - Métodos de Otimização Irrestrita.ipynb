<h3> <strong> 1.1 - Método Gradiente </strong> </h3>

Uma das estratégias mais conhecidas para minimizar uma função é o método clássico do gradiente, também chamado método de Cauchy. É um processo iterativo que à cada etapa faz uma busca na direção oposta ao vetor gradiente da função objetivo no ponto corrente. A justificativa desta escolha se baseia no fato de que, dentre as direções ao longo das quais $f$ decresce, a direção oposta ao gradiente é a de decrescimento mais acentuado. De fato, se $d = \bigtriangledown f(x) $ e $v \in \mathbb{R}^n$ é tal que $ \left \| v \right \| = \left \| d \right \| $ então:
$$
\frac{\partial f }{\partial d}(x) = \bigtriangledown f(x)^{T}d = - 
\left \| \bigtriangledown f(x) \right \|^2 = \left \| \bigtriangledown f(x) \right \| \left \| v \right \| \leq \bigtriangledown f(x)^{T}v = 
\frac{\partial f }{\partial v}(x)
$$
<br>
<strong> Algoritmo 1.1:</strong>  Algoritmo Método Gradiente <p>
Dado: $x^0 \in \mathbb{R}^n$ <br>
$ k = 0 $ <br>
REPITA enquanto $\bigtriangledown f( x^k ) \neq 0 $ <br>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
Defina $d^k = - \bigtriangledown f(x^k)$ <br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
Obtenha $ t_k > 0$ tal que $ f( x^k + t_k d^k ) < f(x^k) $ <br>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
Faça $ x^{k+1} = x^k + t_k d^k $ <br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
$k = k + 1$ <br>

<h3> <strong> 1.2 - Método de Newton </strong> </h3>

O método de Newton é uma das ferramentas mais importantes em otimização. Tanto o algoritmo básico quanto suas variantes são muito utilizados para minimização. <br>

<strong> Motivação </strong> <br>
Considere a função $ f : \mathbb{R}^n \rightarrow \mathbb{R} $ de classe $ C^2 $. Nosso objetivo consiste encontrar um minimizador de $f$. De acordo com as condições necessárias de otimimalidade, devemos resolver o sistema de $n$ equações e $n$ incógnitas dado por $ \bigtriangledown f(x^k) = 0 $. <br>

Generalizando, consiste $F : \mathbb{R}^n \rightarrow \mathbb{R}^n$ de classe $C^1$ e o problema de resolver o sistema(normalmente linear).
$$
F(x) = 0
$$

Como na maioria das vezes não conseguimos resolvê-lo de forma direta, os processos iterativos constituem a forma mais eficiente de lidar com tais situações. <br>

A ideia é aproximar $F$ por seu polinômio de Taylor de primeira ordem. Dada uma estimativa $\bar{x}$, considere o sistema linear.

$$ F( \bar{x} ) + J_F(\bar{x})(x - \bar{x}) = 0 
$$

Onde $J_F$ representa a matriz jacobiana de $F$. Caso $ J_F( \bar{x} ) $ seja inversível, o sistema acima pode ser resolvido, forncendo:
$$
x^+ = \bar{x} - ( J_F(\bar{x}) )^{-1} F(\bar{x})
$$

Voltando agora ao problema de minimizar $f$, aplicamos a estratégia acima para $F = \bigtriangledown f(x)$, obtendo:
$$
x^+ = \bar{x} - (\bigtriangledown^2 f(x))^{-1} \bigtriangledown f( \bar{x} )
$$

<strong> Algoritmo 2.1:</strong>  Algoritmo Método de Newton <p>
Dado: $x^0 \in \mathbb{R}^n$ <br>
$ k = 0 $ <br>
REPITA enquanto $\bigtriangledown f( x^k ) \neq 0 $ <br>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
Defina $d^k = - (\bigtriangledown^2 f(x^k))^{-1} \bigtriangledown f(x^k) $ <br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
Determine o tamanho do passo $t_k > 0$ <br>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
Faça $ x^{k+1} = x^k + t_k d^k $ <br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
$k = k + 1$ <br>


<h3> <strong> 1.3 - Método de Gradiente Conjugado </strong> </h3>

Métodos de direções conjugadas são métodos de primeira ordem (usam apenas informações da função e do gradiente) com convergência mais rápida que o método de Cauchy e custo computacional menor do que Newton. Enquanto Cauchy pode gastar uma infinidade de passos para resolver uma quadrática, Newton a resolve em um passo. Veremos que os métodos de direções conjugadas minimizam uma quadrática definida em $ \mathbb{R}^n $ usando no máximo $n$ passos. <br>

<strong> Algoritmo 2.1:</strong>  Algoritmo Gradiente Conjugado <p>
Dado: $x^0 \in \mathbb{R}^n$ e $d = - \bigtriangledown f(x^0) $<br>
$ k = 0 $ <br>
REPITA enquanto $\bigtriangledown f( x^k ) \neq 0 $ <br>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
$t_k = - \frac{\triangledown f(x^k)^T d^k}{(d^k)^T Ad^k} $

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
Faça $ x^{k+1} = x^k + t_k d^k $ <br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
$ \beta_k = \frac{(d_k)^T A \triangledown f(x{k+1})}{ (d^k)^T A d^k } $

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
$ d^{k+1} = - \triangledown f(x^{k+1}) + \beta_k d^k $

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
$k = k + 1$ <br>


<h3> <strong> 1.4 - Método de Quase-Newton </strong> </h3>

Veremos agora outra classe de métodos que também estão entre Cauchy e Newton no sentido de melhorar a performance em relação ao método do gradiente e ser computacionalmente mais barato quando comparados com Newton. A ideia é construir aproximações para a Hessiana da função objetivo ao longo das iterações. <br>

Assim como no caso de direções conjugadas, os métodos quase-Newton também minimizam uma quadrática em um número finito de passos. <br>

<strong> Algoritmo Básico </strong> <br>
O procedimento iterativo que estudaremos para minimizar uma função $f$ considera as direções de busca dadas por. <br>
$$
d^k = -H \triangledown f(x^k)
$$

Onde $H_k \in \mathbb{R}^{nxn}$ é definida positiva. Tal expressão surge de modo natural quando pensamos, como no caso de newton, em aproximar $f$ por um modelo quadrático em torno de $x^k$. Entretanto, aqui consideramos.
$$
m_k(d) = f(x^k) + \triangledown f(x^k)^T d + \frac{1}{2}d^T B_k d
$$

Onde $b_k \in  \mathbb{R}^{nxn}$ é uma matriz simetrica qualquer ao inves de $\triangledown^2 f(x^k)$. Se $B_k$ for definida positiva, o minimizador do modelo quadrático é dado por:
$$
-B^{-1} \triangledown f(x^k)
$$

<strong> Algoritmo 3.1:</strong>  Algoritmo Método Quase-Newton <p>
Dado: $x^0 \in \mathbb{R}^n$ e $H_0 \in  \mathbb{R}^{nxn} $ definida positiva<br>
$ k = 0 $ <br>
REPITA enquanto $\bigtriangledown f( x^k ) \neq 0 $ <br>


&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
$ d^{k+1} = - H \triangledown f(x^{k})  $

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
Obtenha $t_k > 0 $ que minimiza $f(x^k + td^k)$ em $[0, \infty)$ <br>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
Faça $ x^{k+1} = x^k + t_k d^k $ <br>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
Determine $H_{k+1}$ definida positiva <br> 
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
$k = k + 1$ <br>

```
#Bibliotecas necessárioas
from numpy.linalg import norm, inv
import numpy as np
```

### Rótinas usadas internamente


```
def get_d(fun1, fun2 , x):
    inversa = inv(fun2(x))
    return - inversa@fun1(x)

def aurea(f, x, d, intervalo = -1):
    eps1=1e-5; rho=1; bmax=1e8; # parâmetros
    
    theta1 = (3 - 5**(1/2) ) / 2; theta2 = 1 - theta1; # constantes
    
    a = 0; s = rho; b = 2*rho;
    if intervalo == -1:
        # Fase 1: Obtenção do intervalo [a,b]
        while f(x + d*b) < f(x + d*s) and 2*b < bmax:
            a = s; s = b; b = 2*b;
    else:
        a = intervalo[0]; b = intervalo[1]
    
    # Fase 1: Obtenção do minimo
    u = a + theta1*(b-a); v = a + theta2*(b-a);
    while b - a > eps1:
        if f(x + d*u) <= f(x + d*v):
            b = v; v = u; u = a + theta1*(b-a);
        else:
            a = u; u = v; v = a + theta2*(b-a);
    t = (u+v) / 2
    return t
```

### 1ª Método: Gradiente


```
def gradiente(fun, gradfun ,x0):
    k = 0
    eps1 = 1e-5
    kmax = 10**8
    x = x0
    g = gradfun(x)
    normg = norm(g)
    
    f = gradfun(x)
    
    while normg > eps1 and k < kmax:
        t = aurea(fun, x, -g)
        x = x - t*g
        g = gradfun(x)
        normg=norm(g)
        k=k+1
        f = fun(x)
        
    return {'Ponto Ótimo:': x, 'f(Ótimo)': f}
```

### 2ª Método: Newton



```
def newton(fun, gradfun, hessfun ,x0):
    k = 0
    eps1 = 1e-5
    kmax = 10**4
    x = x0
    g = gradfun(x)
    normg = norm(g)
    
    f = gradfun(x)
    
    while normg > eps1 and k < kmax:
        d = get_d(gradfun, hessfun , x)
        t = aurea(fun, x, d)
        x = x + t*d
        g = gradfun(x)
        normg=norm(g)
        k=k+1
        f = fun(x)
        
    return {'Ponto Ótimo:': x, 'f(Ótimo)': f}
```

### 3ª Método: Gradiente Conjugado HS




```
def gradient_conjugate_HS(f, grad_f, hessian, x, e=1e-3, itermax=10000):
    A = hessian(x)
    d_k = -grad_f(x)
    t_k = aurea(f, x, d_k)
    x = x + t_k*d_k
    k=1
    while np.linalg.norm(grad_f(x)) > e :
        d_k = -grad_f(x) + ((np.dot(grad_f(x).T, np.dot(A, d_k)))/(np.dot(d_k.T, np.dot(A, d_k))))*d_k
        t_k = aurea(f, x, d_k)
        x = x + t_k*d_k
        if k > itermax:
            break
        k += 1
        
    return (f(x), x)
```

### 4ª Método: Quase-Newton




```
### Código
def quase_newton_dfp(fun, fp, x0, maxiter=None, tol=1e-6):

    x0 = np.asarray(x0).flatten()
    if maxiter is None:
        maxiter = len(x0) * 100

    N = len(x0)
    xk = x0
    gk = fp(x0)
    fk = fun(x0)
    Hk = np.eye(N)
    iters = []
    
    for k in range(maxiter):
        pk = - np.dot(Hk, gk)    # compute descent direction
        
        # try line search
        try:
            #alpha_k = _line_search(fun, fp, xk, gk, pk)
            alpha_k = aurea(fun, xk, pk)
        except Exception as e:
            # line search failed
            print('Line search failed: {}'.format(e))
            break

        sk = alpha_k * pk
        xk = xk + sk
        gk_new = fp(xk)
        yk = gk_new - gk
        gk = gk_new
        fk = fun(xk)
        
        # break if norm less than tol
        gknorm = norm(gk)
        if gknorm < tol:
            break


        # update approx of inverse Hessian using Sherman-Morrison formula
        rhok = np.dot(Hk, yk)
        a1 = (sk[:,np.newaxis]*sk[np.newaxis,:]) / np.dot(yk, sk)
        a2 = (rhok[:,np.newaxis]*rhok[np.newaxis,:]) / np.dot(yk, rhok)
        Hk = Hk + a1 - a2
     
    return {'Ponto Ótimo:': xk, 'f(Ótimo)': fk}
```
